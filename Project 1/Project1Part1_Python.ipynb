```
#1 จงเขียนโปรแกรมรับค่า Input จากผู้ใช้ ถ้าผู้ใช้ใส่ 0.0 จึงจะสิ้นสุดโปรแกรม คำนวณหาค่าเฉลี่ยของเลขทั้งหมดที่ผู้ใช้ใส่ และปริ้นค่าเฉลี่ยคูณด้วยจำนวนครั้งที่ใส่ทั้งหมด
sum = 0
n = 0
a = int(input('ใส่ค่า  '))
if a == 0:
  print('ค่าเฉลี่ยคือ ศูนย์')
if a != 0:  
  while a != 0:
    sum = sum + a
    n = n + 1
    a = int(input('ใส่ค่า  '))
  avg = sum/n
  print('ค่าเฉลี่ยคือ {}'.format(avg))
```

```
#2 จงเขียนโปรแกรมให้ผู้ใช้ใส่เลข ขอบเขตล่าง และ ขอบเขตบน และคำนวณหาเลขจำนวนเฉพาะ (Prime Number) ที่อยู่ระหว่างขอบเขตล่างและขอบเขต
upper = int(input('ใส่ค่าขอบเขตบน  '))
lower = int(input('ใส่ค่าขอบเขตล่าง  '))

listN= [2,3,5,7,11,19]

for num in range(lower,upper+1):
  if num > 1:
    for i in range(2,num):
      if (num % i) == 0:
        break
    else:
      print(num)
         

```
