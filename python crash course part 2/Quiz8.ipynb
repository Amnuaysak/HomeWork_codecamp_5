เขียนโปรแกรมให้รับค่าอาหาร 4 ครั้งที่ไปทานอาหารนอกบ้านล่าสุด
โดยแต่ละครั้งต้องไม่ต่ำกว่า 1,000 และไม่เกิน 5,000 (ผู้ใช้ต้องใส่อยู่ใน
ช่วงนี้) และนำค่าอาหาร 3 ครั้งแรกมาบวกกัน หากเกิน 6,000 บาท จะได้
รับส่วนลด 15% หากเกิน 12,000 จะได้รับส่วนลด 25% ของมื้อที่ 4 จง
หายอดเงินสุดท้ายที่ต้องจ่ายของมื้อที่ 4

```
a=float(input('ค่าอาหารครั้งแรก   '))
b=float(input('ค่าอาหารครั้งที่สอง   '))
c=float(input('ค่าอาหารครั้งที่สาม   '))
d=float(input('ค่าอาหารครั้งที่สี่   '))
discountA = 0.15 #ส่วนลดกรณีเกิน 6000 บาทแต่ไม่ถึง 12000 
discountB = 0.25 #ส่วนลดกรณีเกิน 12000 บาท
sum = a+b+c
print('ราคาอาหารรวมทั้งสิ้นเท่ากับ {}'.format(sum))
priceA = 0
priceB = 0
#ผมตีความโจทย์ว่า ส่วนทีเกิน 12000 เท่านั้นที่จะได้ ส่วนลด 25%

if sum>6000 and sum <=12000:
  discount = sum*discountA
  priceA = sum - discount
  total = priceA
if sum>12000:
  discount = d*discountB
  priceB = d - discount
  total =  priceB + sum -sum*discountA
  print('ราคาอาหารครั้งที่ สี่ ที่คิดส่วนลดแล้ว {}'.format(priceB))
print('ราคาอาหารทั้งหมดที่คิดส่วนลดแล้ว {}'.format(total))
```
