เขียนโปรแกรม infinity while loop โดยรับค่า input มาเรื่อยๆ จนกว่า
ผู้ใช้จะใส่เลข 0 และนำเลขทั้งหมดมาหาค่าเฉลี่ย

```
sum = 0
n = 0
a = int(input('ใส่ค่า  '))
if a == 0:
  print('ค่าเฉลี่ยคือ ศูนย์')
if a != 0:  
  while a != 0:
    sum = sum + a
    n = n + 1
    a = int(input('ใส่ค่า  '))
  avg = sum/n
  print('ค่าเฉลี่ยคือ {}'.format(avg))
```
