เขียนโปรแกรมรับ input 4 ตัวที่เป็น string จากนั้นตรวจสอบว่า
input ใดมีความยาวมากที่สุด รองลงมา และน้อยที่สุด จากนั้น ปริ้นแต่ละ
input ออกมาตามลำดับน้อยสุดไปมากสุด

```
lst = []
dictlist = {}
lstkey = []

for i in range(4):
  a = input('ใส่ string ตัวที่ {}'.format(i+1))  
  lst.append(a)
for i in lst:
  dictlist[len(i)] = i
print('ก่อนเรียงลำดับจากความยาวมากไปน้อย',lst)

for key in dictlist:
  lstkey.append(key)

lstkey.sort()
print('เริ่มการเรียงลำดับจำนวน sting จากน้อยไปมาก ดังต่อไปนี้')
for i in lstkey:
  print(dictlist[i])
  

```
