เขียนโปรแกรมรับค่า int มา 3 จำนวน และตรวจสอบว่า 3 เลขนั้น
สามารถนำมาสร้าง 3 เหลี่ยมมุมฉากได้หรือไม่

```
a = float(input('รับค่าความยาวของสามเหลี่ยมด้านที่ 1 '))
b = float(input('รับค่าความยาวของสามเหลี่ยมด้านที่ 2 '))
c = float(input('รับค่าความยาวของสามเหลี่ยมด้านที่ 3 '))
#หาด้านทียาวที่สุด และด้านประกอบสามเหลี่ยมอีกสองด้าน
max = a
sub1 = b
sub2 = c
if b>max:
  max = b
  sub1 = a
  if c>max:
    max = c
    sub2 = b

if (max**2 == sub1**2 + sub2**2):
  print('ความยามสามด้านสร้างสามเหลี่ยมมุมฉากได้')
else:
  print('ความยามสามด้านสร้างสามเหลี่ยมมุมฉากไม่ได้')
```
