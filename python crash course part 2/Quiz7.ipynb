เขียนโปรแกรมให้รับค่าอาหาร 3 ครั้งที่ไปทานอาหารนอกบ้านล่าสุด
โดยแต่ละครั้งต้องไม่ต่ำกว่า 1,000 และไม่เกิน 5,000 (ผู้ใช้ต้องใส่อยู่ใน
ช่วงนี้) และนำค่าอาหารทั้ง 3 ครั้งมาบวกกัน หากเกิน 6,000 บาท จะได้
รับส่วนลด 15% หากเกิน 12,000 จะได้รับส่วนลด 25% จงหายอดเงิน
สุดท้ายที่ต้องจ่าย

```
a=float(input('ค่าอาหารครั้งแรก   '))
b=float(input('ค่าอาหารครั้งที่สอง   '))
c=float(input('ค่าอาหารครั้งที่สาม   '))
discountA = 0.15 #ส่วนลดกรณีเกิน 6000 บาทแต่ไม่ถึง 12000 
discountB = 0.25 #ส่วนลดกรณีเกิน 12000 บาท
sum = a+b+c
print('ราคาอาหารรวมทั้งสิ้นเท่ากับ {}'.format(sum))
priceA = 0
priceB = 0
#ผมตีความโจทย์ว่า ส่วนทีเกิน 12000 เท่านั้นที่จะได้ ส่วนลด 25%

if sum>6000 and sum <=12000:
  discount = sum*discountA
  priceA = sum - discount
  total = priceA
if sum>12000:
  discount = (sum-12000)*discountB
  priceB = (sum-12000) - discount
  total = (12000-12000*discountA) + priceB
print('ราคาอาหารที่คิดส่วนลดแล้ว {}'.format(total))
```
