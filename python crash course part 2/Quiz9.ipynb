เขียนโปรแกรมให้รับค่าอาหาร 4 ครั้งที่ไปทานอาหารนอกบ้านล่าสุด
โดยแต่ละครั้งต้องไม่ต่ำกว่า 1,000 และไม่เกิน 5,000 (เขียนให้ผู้ใช้ใส่
ใหม่หากไม่อยู่ในช่วงที่กำหนด) และนำค่าอาหาร 3 ครั้งแรกมาบวกกัน
หากเกิน 4,000 บาท จะได้รับส่วนลด 25% หากเกิน 9,000 จะได้รับ
ส่วนลด 30% ของมื้อที่ 4 จงหายอดเงินสุดท้ายที่ต้องจ่ายของมื้อที่ 4

```
def price(x):
  while x<1000 or x>5000:
    x=float(input('1000<ค่าอาหาร<5000 .. ป้อนใหม่ '))
  return x  
```

```
a = float(input('ค่าอาหาร '))
a = price(a)

b = float(input('ค่าอาหาร '))
b = price(b)

c = float(input('ค่าอาหาร '))
c = price(c)

d = float(input('ค่าอาหาร '))
d = price(d)

discountA = 0.25 #ส่วนลดกรณีเกิน 4000 บาทแต่ไม่ถึง 9000 
discountB = 0.30 #ส่วนลดกรณีเกิน 9000 บาท

sum = a+b+c
print('ราคาอาหารรวมทั้งสิ้นเท่ากับ {}'.format(sum))
priceA = 0
priceB = 0

if sum>4000 and sum <=9000:
  discount = sum*discountA
  priceA = sum - discount
  total = priceA
if sum>9000:
  discount = d*discountB
  priceB = d - discount
  total =  priceB + sum -sum*discountA
  print('ราคาอาหารครั้งที่ สี่ ที่คิดส่วนลดแล้ว {}'.format(priceB))
print('ราคาอาหารทั้งหมดที่คิดส่วนลดแล้ว {}'.format(total))
```
