เขียนโปรแกรมรับ input สองตัวเป็นจำนวนเต็มระหว่าง 0-60 โดยตัว
แรกเป็นชั่วโมงและตัวที่สองเป็นนาที คำนวณค่าที่จอดรถถ้าชั่วโมงละ
150 บาท เศษนาทีคิดเป็นนาทีละ 2 บาท

```
def intpD(x):
  while x<0 or x>60:
    x=float(input('0<ค่า<60 .. ป้อนใหม่ '))
  return x
```

```
hour = int(input('จำนวนชั่วโมง'))
hour = intpD(hour)
min = int(input('จำนวนนาที'))
min = intpD(min)

fee = hour*150+min*2
print('ค่าที่จอดรถจำนวน {} ชั่วโมง {} นาที เป็นจำนวน {} บาท'.format(hour,min,fee))
```
