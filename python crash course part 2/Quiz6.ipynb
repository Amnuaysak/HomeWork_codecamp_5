เขียนโปรแกรมรับค่า int มา 3 จำนวน และตรวจสอบว่าเมื่อนำ 3 เลข
นั้นมาสร้าง 3 เหลี่ยม จะได้สามเหลี่ยมแบบใด

```
def tellAngle(x):
  a = 'ฉาก'
  b = 'ป้าน'
  c = 'แหลม'

  if x == 90:
    return a
  elif x > 90:
    return b
  if x < 90:
    return c
```

```
import math 

a = float(input('รับค่าความยาวของสามเหลี่ยมด้านที่ 1 '))
b = float(input('รับค่าความยาวของสามเหลี่ยมด้านที่ 2 '))
c = float(input('รับค่าความยาวของสามเหลี่ยมด้านที่ 3 '))
#หาด้านทียาวที่สุด และด้านประกอบสามเหลี่ยมอีกสองด้าน
max = a
sub1 = b
sub2 = c

if b>max:
  max = b
  sub1 = a
  if c>max:
    max = c
    sub2 = b

#ใช้กฏของ cosine ตรวจสอบว่า ความยาวของด้านที่ใส่เข้ามาสร้างเป็นสามเหลียมได้หรือไม่ โดยหาค่ามุมภายในของด้านที่ให้
angleCosMax = math.acos(((sub1**2 + sub2**2) - max**2)/(2*sub1*sub2))
zetaMax = angleCosMax*180/math.pi

angleCosSub1 = math.acos(((max**2 + sub2**2) - sub1**2)/(2*max*sub2))
zetaSub1 = angleCosSub1*180/math.pi

angleCosSub2 = math.acos(((sub1**2 + max**2) - sub2**2)/(2*sub1*max))
zetaSub2 = angleCosSub2*180/math.pi

#ถ้ามุม ที่ได้จากการคำนวนข้างต้นรวมกันแล้ว ไม่เท่ากับ 180 ไม่ถือว่าเป็นสามเหลี่ยม การคำนวนเผื่อทศนิยมไว้ สามตำแหน่ง

if abs((zetaMax + zetaSub1 + zetaSub2) - 180.000) > 0.001:
  print('ด้านสามด้านที่ป้อนเข้ามาสร้างสามเหลี่ยมไม่ได้')
else:
  print('มุมตรงข้ามของด้านที่ยาวที่สุดคือ {} และเป็นมุม {}'.format(zetaMax,tellAngle(zetaMax)))
  print('มุมตรงข้ามของด้านประกอบคือ {} และเป็นมุม {}'.format(zetaSub1,tellAngle(zetaSub1)))
  print('มุมตรงข้ามของด้านประกอบคือ {} และเป็นมุม {}'.format(zetaSub2,tellAngle(zetaSub2)))



```
