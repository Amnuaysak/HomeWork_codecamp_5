เขียนโปรแกรมรับ input สองตัวเป็นจำนวนเต็มระหว่าง 0-60 โดยตัว
แรกเป็นชั่วโมงและตัวที่สองเป็นนาที คำนวณค่าที่จอดรถถ้าชั่วโมงละ
300 บาท เศษนาทีเกิน 15 นาทีคิดเป็นหนึ่งชั่วโมง หากไม่เกินไม่คิดเงิน
และชั่วโมงแรกจอดฟรี

```
def intpD(x):
  while x<0 or x>60:
    x=float(input('0<ค่า<60 .. ป้อนใหม่ '))
  return x
```

```
hour = int(input('จำนวนชั่วโมง'))
hour = intpD(hour)
min = int(input('จำนวนนาที'))
min = intpD(min)


if min>15:
  hour = hour+1

if hour == 0:
  print('จอดฟรี') #จอดฟรี ถ้าเกิน 15 นาทีก็ไม่เสียค่าจอด แต่ถ้าเกิน 15 นาทีก็ปัดเป็น 1 ชั่วโมงแต่ชั่วโมงแรกก็จอดฟรี ดังนั้นไม่เสียค่าจอด
elif hour > 0 and min >15:
  hourCal = hour-1 #หักชั่วโมงแรกออกเพราะจอดฟรี
  hourCal = hourCal+1 #เกิน 15 นาทีปัดเป็นหนึ่งชั่วโมง
  fee = hourCal*300
  print('ค่าที่จอดรถจำนวน {} ชั่วโมง {} นาที เป็นจำนวน {} บาท'.format(hour,min,fee))
if hour > 0 and min <15:
  hourCal = hour-1 #หักชั่วโมงแรกออกเพราะจอดฟรี
  fee = hourCal*300
  print('ค่าที่จอดรถจำนวน {} ชั่วโมง {} นาที เป็นจำนวน {} บาท'.format(hour,min,fee))
```
