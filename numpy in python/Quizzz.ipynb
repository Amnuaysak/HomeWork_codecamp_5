ข้อ 4 จงเขียน array ที่มีเป็นการ random ตัวเลข 20 ตัว แบบ float ค่าอยู่ในช่วง 10-99 แล้วนำแต่ละเลขยกกำลังสอง แต่เขียนแบบ List comprehension

```
a = np.random.randint(10,99,20)
b = [i**2 for i in a]
print(b)
```
